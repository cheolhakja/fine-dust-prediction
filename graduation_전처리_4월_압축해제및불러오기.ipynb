```python
#4월의 미세먼지, 대기질 데이터만 필터링
import pandas as pd
df_air = pd.read_csv('/content/drive/MyDrive/graduation_project/filtered_air_quality_data.csv')

# datetime 컬럼 만들기 (dataTime이 날짜 정보)
df_air['datetime'] = pd.to_datetime(df_air['dataTime'], errors='coerce')
df_air = df_air.dropna(subset=['datetime'])

# 4월 데이터만 추출
df_air_april = df_air[df_air['datetime'].dt.month == 4].copy()
print("✅ 4월 공기질 데이터:", df_air_april.shape)


```

```python
df_air_april.to_csv(
    '/content/drive/MyDrive/graduation_project/air_202504.csv',
    index=False,
    encoding='utf-8-sig'
)
```

```python
#4월 방재기상데이터

import os
from glob import glob
import pandas as pd

base_path = '/content/drive/MyDrive/graduation_project'
weather_files = []

for i in range(1, 10):
    folder = os.path.join(base_path, str(i), 'extracted')
    if os.path.exists(folder):
        weather_files.extend(glob(os.path.join(folder, '*2025-04*.csv')))

print(f"✅ 4월 기상 파일 개수: {len(weather_files)}개")
```

```python
#4월 방재기상데이터 모두 읽고 하나로 합치기
df_weather = pd.concat(
    [pd.read_csv(f, encoding='cp949') for f in weather_files],
    ignore_index=True
)
print("✅ 병합된 행 수:", df_weather.shape)

```

```python
#4월 방재기상데이터 . 컬럼명 확인



print(df_weather.columns.tolist())
```

```python
#4월 방재기상데이터 datetime  변환
df_weather['datetime'] = pd.to_datetime(df_weather['일시'], errors='coerce')
df_weather = df_weather.dropna(subset=['datetime'])
```
