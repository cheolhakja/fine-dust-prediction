```python
df = pd.read_csv('/content/drive/MyDrive/gnn/gnn_2d_202506.csv', encoding='utf8')

feature_cols = [
    'so2Value', 'coValue', 'o3Value', 'no2Value', 'pm10Value', 'pm25Value',
    '풍속(m/s)_1h', '강수량(mm)_1h', '기온(°C)_1h', '풍향(deg)_1h'
]

# 1. 시간/측정소 정렬
df = df.sort_values(['일시', 'stationName'])

# 2. unique 목록
time_list = df['일시'].unique()
station_list = df['stationName'].unique()

#feature_cols 전부를 숫자(float)로 강제 변환
for col in feature_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# adj_station_list: adj행렬 만들 때 사용한 정확한 순서 (리스트)
pivot = df.pivot_table(index='일시', columns='stationName', values=feature_cols)

tensor_list = []
for feat in feature_cols:
    feat_df = pivot[feat][common_stations]
    tensor_list.append(feat_df.values[..., np.newaxis])
tensor3d = np.concatenate(tensor_list, axis=2)
print(tensor3d.shape)
```
